```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# cons
df = pd.read_csv(r"C:\Users\Home\Desktop\CTBA\M3\ConstructionTimeSeriesDataV2.csv", index_col='Month-Year', parse_dates=True)

df
```

```python
fig,ax = plt.subplots(figsize = (20,4))

plt.plot(df['Total Construction'])
plt.suptitle('Total Construction', fontsize=20)
plt.ylabel('Spending')
plt.xticks(np.arange(0,146,12))


# jan
for x in range(0, 146, 12):
    ax.axvline(x, linewidth=1, color='g', linestyle='-.')

# jul
for x in range(6, 146, 12):
    ax.axvline(x, linewidth=1, color='r', linestyle='--')

ax.text(100, 100000, 'Jan in green dashed line, Jul in red dotted line', fontsize=14, bbox=dict(facecolor='white', alpha=0.8))

plt.show()

```

The total construction show a seasonal pattern, low spending in the winter time and high spending during summer time.
